```python
!g++ density.cpp -O3 -ffast-math -march=native
def drawDensity(n, t, mu, c, norm):
    res = !./a.out $n $n $t $mu $c $norm
    
    G_RES = []
    for i in range(0, n):
        G_RES.append([])
        ar = res[i].split(" ")
        for j in range(len(ar)-1):
            G_RES[i].append(float(ar[j]))
            
    G_REAL = []
    for i in range(n, 2*n):
        G_REAL.append([])
        ar = res[i].split(" ")
        for j in range(len(ar)-1):
            G_REAL[i%n].append(float(ar[j]))
            
    V1_RES = []
    for i in range(2*n, 3*n):
        V1_RES.append([])
        ar = res[i].split(" ")
        for j in range(len(ar)-1):
            V1_RES[i%n].append(float(ar[j]))
            
    V1_REAL = []
    for i in range(3*n, 4*n):
        V1_REAL.append([])
        ar = res[i].split(" ")
        for j in range(len(ar)-1):
            V1_REAL[i%n].append(float(ar[j]))
            
            
    V2_RES = []
    for i in range(4*n, 5*n):
        V2_RES.append([])
        ar = res[i].split(" ")
        for j in range(len(ar)-1):
            V2_RES[i%n].append(float(ar[j]))
            
    V2_REAL = []
    for i in range(5*n, 6*n):
        V2_REAL.append([])
        ar = res[i].split(" ")
        for j in range(len(ar)-1):
            V2_REAL[i%n].append(float(ar[j]))
    
    
    
    
    
    f, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(40,40))
    im1 = ax1.pcolormesh(G_RES)
    ax1.set_title("ПОСЧИТАЛ")
    ax2.set_title("РЕАЛЬНО")
    im2 = ax2.pcolormesh(G_REAL)
    
            
    ax1.quiver(V2_RES,V1_RES, color="red",scale=100)
    ax2.quiver(V2_REAL,V1_REAL, color="red",scale=100)
    

    f.colorbar(im1, ax=ax1)
    f.colorbar(im2, ax=ax2)
```

```python
drawDensity(60,10,0.001,10,"L2")
```
